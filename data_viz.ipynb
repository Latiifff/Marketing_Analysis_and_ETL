```python
import matplotlib.pyplot as plt

# Data
ages = [17, 18, 20, 21, 22, 23, 24, 25, 26, 28, 29, 31, 34, 35, 36, 37, 38, 39, 41, 42, 43, 44]
user_count_by_age = [3, 1, 3, 2, 1, 3, 2, 2, 1, 3, 1, 1, 3, 2, 2, 3, 3, 4, 2, 1, 3, 3]

# Plot
plt.figure(figsize=(10, 6))
plt.bar(ages, user_count_by_age, color='green')
plt.xlabel('Age')
plt.ylabel('User Count')
plt.title('User Distribution by Age')
plt.xticks(ages)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
countries = ['America', 'America', 'Canada', 'England', 'France', 'Indonesia', 'Indonesia', 'Singapore']
cities = ['California', 'New York', 'Ottawa', 'London', 'Paris', 'Jakarta', 'Semarang', 'Singapore']
user_count_by_city = [5, 4, 5, 5, 10, 5, 5, 10]

# Plot
plt.figure(figsize=(10, 6))
plt.barh(range(len(user_count_by_city)), user_count_by_city, color='lightgreen')
plt.yticks(range(len(user_count_by_city)), [f"{country} - {city}" for country, city in zip(countries, cities)])
plt.xlabel('User Count')
plt.title('User Distribution by Country and City')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
